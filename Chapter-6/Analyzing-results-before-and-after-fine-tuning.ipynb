```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('df_flan_t5_pretrained_eval.csv')

# Convert the 'Generated Answer' column to lower case to make it case insensitive
df['Generated Answer'] = df['Generated Answer'].str.lower()

# Create a new column 'Answer Category'
df['Answer Category'] = df['Generated Answer'].apply(lambda x: x if x in ['yes', 'no'] else 'Long answer')

# Count the number of 'Yes', 'No', and 'Long answer' in the 'Answer Category' column
answer_category_counts = df['Answer Category'].value_counts()

# Plot the counts
plt.figure(figsize=(8,6))
answer_category_counts.plot(kind='bar')
plt.title('Number of "Yes", "No", and "Long answer" using pre-trained model')
plt.xlabel('Answer Category')
plt.ylabel('Count')
plt.show()
```

```python
# Load the new data
df_new = pd.read_csv('df_flan_t5_fine_tuned_eval.csv')

# Convert the 'Prediction' column to lower case to make it case insensitive
df_new['Prediction'] = df_new['Prediction'].str.lower()

# Create a new column 'Answer Category'
df_new['Answer Category'] = df_new['Prediction'].apply(lambda x: x if x in ['yes', 'no'] else 'Long answer')

# Count the number of 'Yes', 'No', and 'Long answer' in the 'Answer Category' column
answer_category_counts_new = df_new['Answer Category'].value_counts()

# Plot the counts
plt.figure(figsize=(8,6))
answer_category_counts_new.plot(kind='bar')
plt.title('Number of "Yes", "No", and "Long answer" using fine-tuned model')
plt.xlabel('Answer Category')
plt.ylabel('Count')
plt.show()
```
